```python
from scipy.stats import zscore

# Calculate Z-scores for each column
z_scores = zscore(usa_housing.select_dtypes(include='float'))

# Define a threshold for Z-scores (e.g., 3)
threshold = 3

# Filter rows where any Z-score exceeds the threshold
outliers = (z_scores > threshold).any(axis=1)

# Remove outliers
usa_housing = usa_housing[outliers]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

x = usa_housing.drop(['Price'], axis = 1 )
y = usa_housing['Price']
```
