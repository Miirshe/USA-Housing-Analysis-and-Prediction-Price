* **'Mean':** Represents the average value of each variable. It gives you a central measure around which the data tends to cluster.

* **'Std (Standard Deviation)':** Measures the amount of variation or dispersion in the data. A higher standard deviation indicates greater variability.

# Create visualizations for numerical variables using Matplotlib

```python
# defines list numerical variables 
columns =  ['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
                 'Avg. Area Number of Bedrooms', 'Area Population', 'Price']
# defines list colors of each variable
colors = [
    "lightcoral","lightgreen","lightpink","lightpink","lightblue","orange"
]

# set up subplot
fig ,axes = plt.subplots(nrows=2, ncols = 3 , figsize = (15,8))
fig.suptitle ('Distributions of Numerical Variables')

for i , (col , color ) in enumerate(zip(columns , colors)):
    axes[i // 3 , i % 3].hist(usa_housing[col], bins=20, color=color, edgecolor='black')
    axes[i // 3 , i % 3].set_title(col)
    axes[i // 3 , i % 3].set_xlabel(col)
    axes[i // 3 , i % 3].set_ylabel('Frequency')
    
# Adjust layout
plt.tight_layout(rect=[0, 0.03, 1, 0.95])

# show plots    
plt.show()
```

```python
# Pairplot displays for numeric variables
sns.pairplot(usa_housing)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 2)
plt.hist(usa_housing['Price'], bins=20, color=color, edgecolor='black')
plt.xlabel('before log transform ')  # Corrected line
plt.title('Histogram before log transform')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
from scipy.stats import zscore

# Calculate Z-scores for each column
z_scores = zscore(usa_housing.select_dtypes(include='float'))

# Define a threshold for Z-scores (e.g., 3)
threshold = 3  

# Filter rows where any Z-score exceeds the threshold
outliers = (z_scores > threshold).any(axis=1)

# Remove outliers
usa_housing = usa_housing[outliers]
data =  usa_housing.drop(['Address'], axis = 1 )
plt.figure(figsize = (15,8))
sns.heatmap(data.corr(), annot = True , cmap = "YlGnBu")
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 2)
plt.hist(usa_housing['Price'], bins=20, color=color, edgecolor='black')
plt.xlabel('before log transform ')  # Corrected line
plt.title('Histogram before log transform')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

x = usa_housing.drop(['Price','Address'], axis = 1 )
y = usa_housing['Price']
```

```python
x
```

```python
y
```

```python
# Split the data into training and testing sets
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.2)
```

```python
data_train = x_train.join(y_train)
```

```python
data_train
```

```python
data_train.hist(figsize = (15,8))
```

```python
plt.figure(figsize = (15,8))
sns.heatmap(data_train.corr(), annot = True , cmap = "YlGnBu")
```

```python
usa_housing['Address'].value_counts()
```

```python
data_train['Avg. Area Number of Rooms'] = np.log(data_train['Avg. Area Number of Rooms'] + 1)
data_train['Avg. Area Number of Bedrooms'] = np.log(data_train['Avg. Area Number of Bedrooms'] + 1)
data_train['Area Population'] = np.log(data_train['Area Population'] + 1)
```

```python
data_train.hist(figsize = (15,8))
```

```python
data_train['bedromm_ratio'] = data_train['Avg. Area Number of Bedrooms'] / data_train['Avg. Area Number of Rooms'] 
```

```python
plt.figure(figsize = (15,8))
sns.heatmap(data_train.corr(), annot = True , cmap = "YlGnBu")
```

```python
# Initialize the linear regression model
model = LinearRegression()

# Train the model
model.fit(x_train, y_train)

# Make predictions on the test set
y_pred = model.predict(x_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
```
